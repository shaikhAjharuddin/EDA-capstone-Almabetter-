#### Chart - 1
# Which countries in the world is least effected by terrorism? 


````python
# Chart - 1 

try:
  country_wise_attack_count = terror_master_data ['Country'].value_counts()         #Counting number of times each country name appears
  country_wise_attack_count.sort_values(axis =0 , inplace=True,ascending=[True] )   #Sorting countries as per no. of occurances
  print(type(country_wise_attack_count))
except Exception as e:
  print(e)
else:
  # Plotting bar graph for 20 countries
  plt.rcParams['figure.figsize']=(15,8)
  country_wise_attack_count[20::-1].plot(kind='line', color = "coral", linewidth = 10)
  plt.title('Total terror attacks country wise')
  

````

##### 1. Why did you pick the specific chart?

Bar graphs provide a very **easy perception**. Humans have a better **understanding with length differences** than areas or angles. Hence, as we wanted to compare the attacks in different countries, we chose bar graph. Horizontal bar was chosen to accomodate more countries in the screen area.

##### 2. What is/are the insight(s) found from the chart?

The chart clearly shows those countries which are least effected by terrorism. We inferred that the common denominator between countries like North Korea, Antigua & Barboda, Vatican city was the fact that these countries **do not have very significant religious diversities** which may lead to dissent. #### Chart - 9
Analyse the terror attacks on the cities of Jammu and Kashmir.

````python
# Chart - 10 visualization code

try:
  #Filtering records of India
  terror_data_JK =terror_master_data[terror_master_data['state'] == 'Jammu and Kashmir']
  #Grouping cities and sorting them by number of successful terrorist attacks
  terror_data_JK_cities = terror_data_JK.groupby('city')['Success'].sum()
  terror_data_JK_cities.sort_values(axis =0 , inplace=True, ascending=[False])
  terror_data_JK_cities = terror_data_JK_cities.drop("Unknown")
except Exception as e:
  print(e)
else:
  #Plotting graph to show 20 majir terrorist targets in India 
  sns.barplot(x= terror_data_JK_cities[0:10].index,y = terror_data_JK_cities[0:10].values)

````

#### Chart - 10 - Correlation Heatmap
Analyse the data correlation between various attributes from the Global Terrorism Database.

````python
# Correlation Heatmap visualization code

def findDataCorr():
  # Correlation Heatmap visualization code
  plt.figure(figsize=(15,10))
  #This shows how much related is one parameter to the other in the dataset.
  try: 
    sns.heatmap(np.round(terror_master_data.corr(),2),annot=True, cmap='Pastel2')
  except Exception as e:
    print(e)


findDataCorr()
````
